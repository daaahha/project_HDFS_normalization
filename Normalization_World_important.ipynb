```python
import pandas as pd
from hdfs import InsecureClient
import numpy as np
```

```python
World_Important_Dates = pd.read_csv('World Important Dates.csv', delimiter=',')
```

```python
World_Important_Dates.info()
```

```python
World_Important_Dates.head(10)
```

```python
World_Important_Dates['Date'] = World_Important_Dates['Date'].replace('Unknown', pd.NA)
World_Important_Dates['Date'] = pd.to_numeric(World_Important_Dates['Date'], errors='coerce').fillna(0).astype(int)
World_Important_Dates.insert(World_Important_Dates.columns.get_loc('Year') + 1, 'Era', World_Important_Dates['Year'].apply(lambda x: 'BC' if 'BC' in x else 'AD'))
World_Important_Dates['Year'] = World_Important_Dates['Year'].str.replace('BC', '')
World_Important_Dates['Year'] = World_Important_Dates['Year'].astype(int)
World_Important_Dates['CombinedDate'] = World_Important_Dates['Year'].astype(str) + '-' + World_Important_Dates['Month'].astype(str) + '-' + World_Important_Dates['Date'].astype(str)
World_Important_Dates['CombinedDate'] = World_Important_Dates['CombinedDate'].replace('-Unknown-0', '', regex=True)
```
