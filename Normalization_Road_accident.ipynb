```python
Road_accident_Data['Accident Date'] = pd.to_datetime(Road_accident_Data['Accident Date'], format='%m/%d/%Y')
Road_accident_Data['Carriageway_Hazards'] = Road_accident_Data['Carriageway_Hazards'].fillna('no obstacles')
Road_accident_Data['Latitude'] = Road_accident_Data['Latitude'].round(2)
Road_accident_Data['Longitude'] = Road_accident_Data['Longitude'].round(2)
```
